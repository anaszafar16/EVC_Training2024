```python
class_names = df2['GradeClass'].unique()
df_sampled = [] # Change to a list

for class_name in class_names:
    sampled_rows = df2[df2['GradeClass'] == class_name].head(75)
    df_sampled.append(sampled_rows) # Append to the list

df_sampled = pd.concat(df_sampled, ignore_index=True)
```

```python
df_sampled = df_sampled.drop(df.index[-1])
df_sampled
```

```python
merged_df = pd.concat([df, df_sampled], axis=1)
merged_df
```

```python
merged_df.isnull().sum()
```

```python
merged_df = merged_df.dropna()
merged_df
```

```python
merged_df.columns
```

```python
merged_df = merged_df.drop(['Person ID', 'StudentID','Occupation','Daily Steps','Blood Pressure','Stress Level' ,'Heart Rate','Tutoring','Ethnicity','ParentalEducation','Extracurricular','Sports', 'Music', 'Volunteering','GPA'], axis=1)
merged_df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
merged_df["BMI Category"] = encoder.fit_transform(merged_df["BMI Category"])
merged_df["Sleep Disorder"] = encoder.fit_transform(merged_df["Sleep Disorder"])
```

```python
matrix = merged_df.corr()
matrix
```

```python
X = merged_df.drop("Sleep Duration", axis=1)
y = merged_df["Sleep Duration"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

reg_model = LinearRegression()
reg_model.fit(x_train, y_train)
y_pred = reg_model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
score = reg_model.score(x_train, y_train)
score
```

```python
import joblib
joblib.dump(reg_model, 'regression_model.joblib')
```
